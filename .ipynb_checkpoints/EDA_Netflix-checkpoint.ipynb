# Data Cleaning & Transformation

```python
df_netflix.drop(columns= 'director',axis  = 1,inplace=True)
df_netflix.dropna(subset=['country','cast','date_added'],inplace=True)
rating_mode = df_netflix['rating'].mode().iloc[0]
df_netflix['rating'] = df_netflix['rating'].fillna(rating_mode)
df_netflix.isnull().sum().sort_values(ascending=False)
```

# Movies

```python
df_netflix_movies = df_netflix[df_netflix['type'] == 'Movie']
df_netflix_movies.head(5)
df_netflix_movies.isnull().sum().sort_values(ascending=False)
```

```python

(df_netflix_movies[
        df_netflix_movies['country'].map(df_netflix_movies['country'].value_counts()) > 75]['country'].value_counts().plot(
                kind='pie',
                figsize=(7, 7),
                title='Country Distribution (Count > 75)'))

plt.ylabel('')
plt.show()
```

```python
print(df_netflix_movies['rating'].value_counts())
df_netflix_movies['rating'].value_counts().plot(kind = 'pie',ylabel = 'Ratings',figsize = (5,5))
plt.show()
```

```python
genre_dict = {}

for genres in df_netflix_movies['listed_in']:
    for genre in genres.split(','):
        genre = genre.strip()
        genre_dict[genre] = genre_dict.get(genre, 0) + 1
genre_dict
df_movie_genre = pd.DataFrame.from_dict(genre_dict, orient="index", columns=["count"])
# mean = df_movie_genre['count'].mean()
df_movie_genre[df_movie_genre['count']>df_movie_genre['count'].mean()].plot(kind = "pie", y='count',
                   figsize = (5,5))
```

```python
df_genre_year = (
    df_netflix_movies
    .assign(genre=df_netflix_movies['listed_in'].str.split(', '))
    .explode('genre')
    .groupby(['release_year', 'genre'])
    .size()
    .reset_index(name='count')
)
top_genres = (
    df_genre_year.groupby('genre')['count']
    .sum()
    .sort_values(ascending=False)
    .head(6)
    .index
)

df_top = df_genre_year[df_genre_year['genre'].isin(top_genres)]
plt.figure(figsize=(10,6))
sns.lineplot(data=df_top, x='release_year', y='count', hue='genre')
plt.title("Trend of Top TV Show Genres Over Time")
plt.show()
```

```python
df_netflix_movies['duration_min'] = df_netflix_movies['duration'].str.split(expand=True)[0]
df_netflix_movies
df_netflix_movies.drop('duration', axis = 1)
```

```python
df_netflix_movies['duration_min'].value_counts().plot(ylabel = 'No. of Movies')
```

### ðŸŽ¬ Movies â€“ Summary Insights

Movies constitute the **largest share of Netflixâ€™s content library**, showing a clear emphasis on scalable and high-volume content production. The catalog is dominated by genres such as **Drama, Comedy, and International Movies**, indicating Netflixâ€™s focus on emotionally engaging and globally adaptable storytelling. Rating-wise, movies are heavily concentrated in **TV-MA and TV-14**, reflecting a strategic targeting of **mature and young-adult audiences**, while traditional film classifications like **PG, PG-13, and R** remain almost exclusive to the movie format. Most movies fall within the **80â€“120 minute duration range**, suggesting a preference for standard-length films that align well with binge-friendly viewing habits. The sharp increase in movie additions after **2016** highlights Netflixâ€™s aggressive global expansion phase, during which films played a key role in rapidly growing and diversifying the platformâ€™s content library.

# Estimated Number of Movies in Future Years (Trend-Based)

```python
df_netflix_per_year['Movie'].plot(kind = 'line')
sns.lineplot(x = x,y=y)
```

```python
# df_netflix_per_year = df_netflix_per_year.reset_index()
df_netflix_per_year.dropna(inplace = True)
y = df_netflix_per_year['Movie'] #dependent
x = df_netflix_per_year[['release_year']] #independent (df)
lm = linear_model.LinearRegression() # creates and object of LinearRegression class
lm.fit(x,y)
```

```python
year = [2022,2023,2024,2025]
lm.predict(x)
print("Score: ",round(lm.score(x,y),2))
print("Coefficient: ",round(lm.coef_[0],2))
print("Intercept: ",round(lm.intercept_,2))
year_2025 = pd.DataFrame({'release_year':year})
prediction_2025 = lm.predict(year_2025)
for i in range(0,4):
    print("Predicted movies in "+str(2022+i), int(prediction_2025[i]))
```

# Estimated Number of TV Shows in Future Years (Trend-Based)

```python
# df_netflix_per_year = df_netflix_per_year.reset_index()
df_netflix_per_year.dropna(inplace = True)
y = df_netflix_per_year['TV Show'] #dependent
x = df_netflix_per_year[['release_year']] #independent (df)
lm = linear_model.LinearRegression() # creates and object of LinearRegression class
lm.fit(x,y)
df_netflix_per_year['TV Show'].plot(kind = 'line')
sns.lineplot(x = df_netflix_per_year['release_year'],y=y)
```

```python
year = [2022,2023,2024,2025]
lm.predict(x)
print("Score: ",round(lm.score(x,y),2))
print("Coefficient: ",round(lm.coef_[0],2))
print("Intercept: ",round(lm.intercept_,2))
year_2025 = pd.DataFrame({'release_year':year})
prediction_2025 = lm.predict(year_2025)
for i in range(0,4):
    print("Predicted TV Shows in "+str(2022+i), int(prediction_2025[i]))
```

## ðŸ“Œ Conclusion & Insights

### Why Linear Regression Is Weak in This Context
Linear regression assumes a constant rate of change over time. However, Netflixâ€™s content release strategy does not follow a uniform yearly increase. The platform experienced rapid expansion during certain periods (especially post-2015), followed by stabilization and slight declines in recent years. As a result, linear regression oversimplifies the trend and should be interpreted as a **baseline indicator of long-term direction**, not as an accurate forecasting model.

---

### Why Netflixâ€™s Growth Is Non-Linear
Netflixâ€™s growth is driven by multiple external and strategic factors, including:
- Rapid global expansion during its scaling phase  
- Increased investment in original content during specific years  
- Market saturation and strategic cost optimization in later years  
- External disruptions such as the COVID-19 pandemic  

These factors introduce accelerations, slowdowns, and plateaus in release counts, making the overall trend **non-linear** and unsuitable for simple straight-line modeling.

---

### What a Decision-Maker Should Learn
- Netflix has transitioned from aggressive expansion to **more controlled and selective content releases**.
- TV Shows exhibit a stronger long-term growth trend compared to Movies, indicating a strategic shift toward serialized content.
- Regression-based extrapolations should be used only for **high-level trend understanding**, not operational forecasting.
- Strategic decisions would require more robust approaches such as time-series models and additional business-related features.

---

### Final Note
This analysis demonstrates how exploratory data analysis combined with simple regression techniques can uncover meaningful trends while also highlighting the importance of understanding model limitations. Future work could extend this study using time-series forecasting methods or multi-feature models to better capture Netflixâ€™s evolving content strategy.
